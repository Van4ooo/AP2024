```python
from AP2024.sorts import PyCombSort, PyIntroSort, PySmoothSort
from AP2024 import Measurement, Diagram, D, SortParam
```

```python
# we measure sorting algorithms
count_m = 50_000_000

Measurement(PySmoothSort, max_c=count_m)
Measurement(PyIntroSort, max_c=count_m)
Measurement(PyCombSort, max_c=count_m)
```

```python
# may draw many diagrams with (n*log(n) line) on one diagram
Diagram(
    SortParam.time,
    D([PyIntroSort, PySmoothSort, PyCombSort], 'rand',),
    draw_log=True
)
```
